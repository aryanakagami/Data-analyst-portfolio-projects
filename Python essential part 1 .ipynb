# Comparison , logical and Membership operaator in python

```python
# Comparison operator: Use to compare 
# ==
print(1,10==10)
print(2,10==50)
x = 'vanilla'
y = 'chocolate'
print(3,x==y)
# !=
print(4,10!=10)
print(5,10!=50)
print(6,x!=y)
# <
print(7,10<10)
print(8,10<50)
# >
print(9,10>10)
print(10,10>50)
```

```python
# Logical operators: for logical ans
# and
print(1,(10 > 50) and (50 > 10))
print(2,(70 > 50) and (50 > 10))
# or
print(3,(10 > 50) or (50 > 10))
print(4,(70 > 50) or (50 > 10))
# not
print(5,not(50>10))
```

# While loop

```python
# while loop : use to iterate as long as text condition is true . It need condition.
number = 1
while number<=10:
    print(number)
    number = number + 1
```

```python
number = 1
while number<=10:
    print(number)
    if number == 3:
        break          # with break statement , we can stop the loop even when while condition is true
    number = number + 1
```

```python
number = 1
while number<=10:
    print(number)
    if number == 11:
        break          
    number = number + 1
else:   # works when condition of while statement fallen
    print(' no longer small')
```

```python
# continue statement : if it triggers , then it rejects all remaining statements in current iteration of loop and we go to next iteration
number = 0
while number<=10:
    number = number + 1
    if number == 7:
        continue    
    print(number)
else:   # it rejected 7
    print(' no longer small')
```

# Merge , Join and Concatenate

```python
df1 = pd.read_csv(r"D:\gami\LOTR (1).csv")
df2 = pd.read_csv(r"D:\gami\LOTR 2 (1).csv")
```

```python
print(df1)
print('\n')
print(df2)
```

```python
# merging 1st dataframe to 2nd
df1.merge(df2)
```

```python
# we did'nt explicitly ssays join using fellowshipID
df1.merge(df2,how = 'inner') # inner is default
```

```python
df1.merge(df2 , how = 'inner' , on = 'FellowshipID') 
# if we does'nt join them on basis of firstname , then it will give two diff columns
```

```python
df1.merge(df2 , how = 'inner' , on = ['FellowshipID','FirstName']) 
```

```python
df1.merge(df2 , how = 'outer' ) # their are left and right joins too
```

```python
# their is also a cross join
df1.merge(df2 , how = 'cross')
```

```python
# let see join function
# merge works better with columns
# joins works better with indexes
# df1.join(df2) :  shows error , cannot distinguished which is which (['FellowshipID', 'FirstName'])
df1.join(df2 , on = 'FellowshipID' , how = 'outer' , lsuffix  = '_left',rsuffix = '_right')
```

```python
# doing two steps at a time , joining and setting index
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),lsuffix  = '_left',rsuffix = '_right')
```

```python
df4
```

```python
# concatenate : putting one dataframe above other
pd.concat([df1,df2])
```

```python
# can also do inner , outer ... etc joins
pd.concat([df1,df2] , join = 'inner')
```

```python
# can also change based on axis
pd.concat([df1,df2] , join = 'inner' , axis = 1)
```
