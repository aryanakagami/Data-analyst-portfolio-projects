```python
# EDA: First look to the data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\gami\world_population (2).csv")
```

```python
# displaying into float format with 2 decimals
pd.set_option('display.float_format' , lambda x: '%.2f'%x)
```

```python
# Basic level information of dataset
df.info()
```

```python
# Describing dataset into a big 
df.describe()
```

```python
# seeing how many null value exists null
df.isnull().sum()
```

```python
# now we want to see how many unique values in each of these columns
df.nunique()
```

```python
# we want to see top range of countries
df.sort_values(by = '2022 Population' , ascending=False).head(11)
```

```python
# Now we will look into correlations
df78 = df.corr(numeric_only = True)
```

```python
sns.heatmap(df78, annot = False)
plt.show()
plt.figure(figsize=(10, 8));
```

```python
# grouping by continent will be useful as it will make it easy to analyze
df2 = df.groupby(by = 'Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by='2022 Population' , ascending = False)
df2
```

```python
# I further want to know about ocenia
df[df['Continent'].str.contains('Oceania')]
# after analyzing , we see that Oceania is nothing but Australia old name
```

```python
# Transposing matrix
df3 = df2.transpose()
df3
```
