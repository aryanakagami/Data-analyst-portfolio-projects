# Data Cleaning Step by Step

```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\gami\Customer Call List.xlsx")
df
```

```python
# By getting first glance in our dataframe we can see that no data cleaning required in customer id 
# and first name
# So let's start cleaning by changing last name a little bit but first , we have to remove duplicates
df = df.drop_duplicates()
df
```

```python
# As we can see , anakin skywalker is cleaned up as duplicate
# Now , let's delete not very useful column
df = df.drop(columns = 'Not_Useful_Column')
# The not so useful column was dropped from the data frame
```

```python
df # we are not using inplace so that we will not make any mistakes
```

```python
# now let's move to last name column
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
# this is the best way to do that
# note :- we cannot use df['Last_Name'].str.strip(['...','/']) because it's not like a list
# It's regular expression
df
```

```python
# Now , removing the errors from phone number column
# we just want to gneralize it like 123-545-5421	
# Now , for this , we have to remove all non numeric stuffs from the numbers and add -- after that
# First we will make everything numeric
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True) # ^ it saying that it will replace any character 
# except a-zA-Z0-9 and put '' in it's space
# Also , add regex( regular expression) is to be set as true
# df.replace use to replace from first value to second
df
```

```python
# Now we have to add -- in betweens of number
# Note :- lambda function usage
#Regular function
#    def add(x, y):
#        return x + y

# Equivalent lambda function
# add_lambda = lambda x, y: x + y
# It will show error : df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:])
# 'float' object is not subscriptable
# Because their is no strings
# So, we have to convert this columns into string
# Now , we cannot use str(df['Phone_Number']) , it will add extra nuance
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))
df
```

```python
# Now , we can add -- into phone number
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] + '-' + x[6:])
# After that , remove nans
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].replace('nan--' , '' , regex = True)
df['Phone_Number']= df['Phone_Number'].replace('Na--' , '' , regex = True)
df['Phone_Number']= df['Phone_Number'].replace('--' , '' , regex = True)
df
```

```python
# Now , the phone_number column is also cleaned up
# Now , let's split the column address
df[['Street_Address' , 'State' , 'Zip']]=df['Address'].str.split(',' ,n=2,expand = True) # we wanna split on comma and then how many values left to right it should look for.
# Expand helps to split into different columns
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes' ,'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No' ,'N')
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('Yes' ,'Y')
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('No' ,'N')
```

```python
# for removing na
df.fillna('')
```

```python
# for removing N/A
df['Address'] = df['Address'].str.replace('N/a' , '')
df['Paying Customer'] = df['Paying Customer'].str.replace('N/a' , '')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N/a' , '')
df['Street_Address'] = df['Street_Address'].str.replace('N/a' , '')
```

```python
df
```

```python

# we need to remove those who don't have numbers or who don't want to get called
for x in df.index: # we are doing by index 
    if df.loc[x,'Do_Not_Contact'] == ('Y' or ''):
        df.drop(x , inplace = True)
```

```python
df
```

```python
# for phone number also 
for x in df.index: # we are doing by index 
    if df.loc[x,'Phone_Number'] == ( ''):
        df.drop(x , inplace = True)
```
